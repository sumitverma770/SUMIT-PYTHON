```python
import pandas as pd
import numpy as np
import seaborn as sns
import plotly.express as px

```

```python
iris = sns.load_dataset('iris')
iris 
```

```python
sns.relplot(data=iris, x='sepal_length', y='sepal_width', hue='species')
```

```python
sns.histplot(data=iris, x='petal_length')
```

```python
sns.jointplot(data=iris,x='sepal_length',y='sepal_width',hue='species')
```

```python
sns.FacetGrid(data=iris,col='species',hue='species').map(sns.swarmplot,'sepal_length')
```
